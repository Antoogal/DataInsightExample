libraries used/needed

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('sales_dataset.csv')
df
```

Visualization in python

```python
plt.figure(figsize = (10,6))
sns.regplot(data = df, x = 'marketing_budget(thousands)', y = 'actual_sales(millions)',
            scatter_kws={'color':'green'})
plt.title('Correlation between Marketing Budget and Actual Sales', fontsize = 16, fontweight = 'bold')
plt.grid(False)
plt.show()
```

Conclusion

A linear regression analysis between marketing budget and actual sales reveals a strong positive correlation. Indicating that increasing the marketing budget, increases actual sales.
